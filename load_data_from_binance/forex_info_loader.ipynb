```python
import tradermade as tm
import pandas as pd
import numpy as np
currenciesDataframe = []
for currency in selectedCurrencies:
    manuallyTimeseries = []
    for year in range (2019, 2023):
        print(year)
        timeseriesData = tm.timeseries(currency=currency, start=str(year)+"-01-01",end=str(year+1)+"-01-01",
                                       interval="daily",fields=["high","open","low","close"])
        timeseriesData = timeseriesData.assign(symbol= currency)
        print(timeseriesData)
        manuallyTimeseries.append(timeseriesData)
      
    manuallyTimeseries = pd.concat(df for df in manuallyTimeseries )
    currenciesDataframe.append((manuallyTimeseries)) 

```
